```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow  # for image display
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
```

```python
# Membuat list untuk menyimpan url dari beberapa image
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg","https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg", "https://i.pinimg.com/736x/0b/2d/ae/0b2dae8928e178f368b6d152a1367b7f.jpg"]

# Baca dan tampilkan image
# loop pada tiap url image, beberapa image dapat disimpan pada list
for url in urls:
    image = io.imread(url)                           # read image
    image = cv.resize(image, (0, 0), fx=0.5, fy=0.5) # resize image to half size
    image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)   # convert color to RGB
    final_frame = cv.hconcat([image, image_2])       # concatenate image
    cv2_imshow(final_frame)                          # show image
    print('\n')
```

```python
tinggi = image_2.shape[0]
lebar = image_2.shape[1]

print("resolusi image: tinggi x lebar = ", tinggi, " x ", lebar)

cv2_imshow(image_2)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# membuat garis horizontal di tengah image
for y in range(lebar):
    image_3[int((tinggi)/2), y] = [255, 255, 255]  # putih (R=255, G=255, B=255)

final_frame = cv.hconcat((image_2, image_3))  # gabungkan dua gambar
cv2_imshow(final_frame)
```

```python
image_4 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_5 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_6 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# membuat garis horizontal di tengah image

for y in range(tinggi):
    image_5[y, int((lebar)/2)] = [255, 255, 255]

for x in range(lebar):
    y = int((tinggi/lebar) * x)
    image_6[y, x] = [255,255,255]

for x in range(lebar):
    y = tinggi - 1 - int((tinggi/lebar) * x)
    image_6[y, x] = [255,255,255]

final_frame1 = cv.hconcat((image_4, image_5))
cv2_imshow(final_frame1)

final_frame2 = cv.hconcat((image_4, image_6))
cv2_imshow(final_frame2)
```

```python
image_hor = cv.cvtColor(image, cv.COLOR_BGR2RGB)  # copy gambar

# koordinat tengah secara vertikal
y = int(tinggi*0.70)

# tentukan panjang garis (misal 100 pixel)
start_x = 50
end_x = start_x + 90

# buat garis horizontal putih
for x in range(start_x, end_x):
    image_hor[y, x] = [255, 255, 255]

# tampilkan hasil
final_frame = cv.hconcat((image_4, image_hor))
cv2_imshow(final_frame)
```

```python
image_fill = cv.cvtColor(image, cv.COLOR_BGR2RGB)  # copy gambar

# tentukan posisi kotak (pojok kiri atas dan kanan bawah)
x1, y1 = 40, 40     # pojok kiri atas
x2, y2 = 140, 140    # pojok kanan bawah

# isi penuh kotak dengan warna putih
for y in range(y1, y2):
    for x in range(x1, x2):
        image_fill[y, x] = [255, 255, 255]  # putih

# tampilkan hasil
final_frame = cv.hconcat((image_4, image_fill))
cv2_imshow(final_frame)
```
